```python
# 파일명 불러오기
File_Name = []

# 해당 경로에 해당하는 파일명 불러오기
for i in RCP_Title['RCP_NM']:
    File_Name.append(os.path.basename(f'./Keyword_Count/키워드사운드_{i}_검색량.xlsx'))

File_Name
```

```python
import pandas as pd
import glob


try:
    path = './Keyword_Count/'
    RCP_Count = pd.DataFrame()
    for file_name in File_Name:
        df = pd.read_excel(path+file_name)
        RCP_Count = RCP_Count.append(df, ignore_index=True)
    print(RCP_Count)

except Exception as ex:
    print('error' + str(ex))
```

## 성별 전처리

### Merge

```python
import pandas as pd
import os
import openpyxl as xls
import glob

file_format = ".xlsx"
file_path = 'D:/vsc_project/machinelearning_study/Project/searchData/Data/Male/60'
file_name_os = os.listdir(file_path)
# # print(file_name_os)

file_list = [f"{file_path}/{file}" for file in os.listdir(file_path) if file_format in file]
# print(file_list)

for i in range(len(file_name_os)):
    wb = xls.load_workbook(file_list[i])
    ws = wb.active
    ws.delete_rows(1, 6)
    wb.save(file_list[i])

for i in range(len(file_name_os)-1):
    wb = xls.load_workbook(file_list[i])
    ws = wb.active
    ws.delete_cols(1)
    wb.save(file_list[i])

df = pd.DataFrame()
for i in range(len(file_name_os)):
    df = pd.concat([df, pd.read_excel(file_list[i])], axis = 1)
df.to_excel(file_path + '/' + 'datalabAll' + '.xlsx')
```

```python
import pandas as pd
import os
import openpyxl as xls
import glob

file_format = ".xlsx"
file_path = 'D:/vsc_project/machinelearning_study/Project/searchData/Data/Keyword_Count'
file_name_os = os.listdir(file_path)
# # print(file_name_os)

file_list = [f"{file_path}/{file}" for file in os.listdir(file_path) if file_format in file]
# print(file_list)

for i in range(len(file_name_os)):
    wb = xls.load_workbook(file_list[i])
    ws = wb.active
    ws.delete_cols(3, 2)
    wb.save(file_list[i])

for i in range(len(file_name_os)):
    wb = xls.load_workbook(file_list[i])
    ws = wb.active
    ws['C1'].value = ws['B2'].value
    wb.save(file_list[i])

for i in range(len(file_name_os)):
    wb = xls.load_workbook(file_list[i])
    ws = wb.active
    ws.delete_cols(2)
    wb.save(file_list[i])

for i in range(len(file_name_os)-1):
    wb = xls.load_workbook(file_list[i])
    ws = wb.active
    ws.delete_cols(1)
    wb.save(file_list[i])

df = pd.DataFrame()
for i in range(len(file_name_os)):
    df = pd.concat([df, pd.read_excel(file_list[i])], axis = 1)
df.to_excel(file_path + '/' + 'datalabAll' + '.xlsx')


```
